**3.1**

````python
## Code here

## Get x and y variables
time_train = taxi_train['TimeMin']
pickups_train = taxi_train['PickupCount']

# Add column of 1
X = sm.add_constant(time_train)

# build the OLS model (ordinary least squares) from the training data
OLSModel = sm.OLS(pickups_train, X)

# fit and save regression info (parameters, etc) in results_sm
results_sm = OLSModel.fit()

# pull the beta parameters out from results_sm
beta0 = results_sm.params[0]
beta1 = results_sm.params[1]

print("(beta0, beta1) = (%f, %f)" %(beta0, beta1))
````

**3.2**

````python
## Code for your plot here

## Plot Pickups Over Time
plt.plot(time_train, pickups_train,'yo', alpha=.5, label='Training Data')
plt.title('Taxi Pickups vs Time of Day (Training Set)')
plt.xlabel('Time of Day in Minutes')
plt.ylabel('Number of Taxi Pickups')

# Plot OLS Line
x1 = [0, max(time)]
y1 = [beta0, beta0 + max(time)*beta1]
plt.plot(x1, y1, 'k-', lw=3, label='OLS Line')

# Show Plot
plt.legend()
plt.show()
````

````python
## Plot Pickups Over Time

## Get x and y variables
time_test = taxi_test['TimeMin']
pickups_test = taxi_test['PickupCount']


plt.plot(time_test, pickups_test,'ro', alpha=.5, label='Test Data')
plt.title('Taxi Pickups vs Time of Day (Test Set)')
plt.xlabel('Time of Day in Minutes')
plt.ylabel('Number of Taxi Pickups')

# Plot OLS Line
x1 = [0, max(time)]
y1 = [beta0, beta0 + max(time)*beta1]
plt.plot(x1, y1, 'k-', lw=3, label='OLS Line')

# Show Plot
plt.legend()
plt.show()
````

````python
plt.plot(time_test, pickups_test,'ro', alpha=1, label='Test Data')
plt.plot(time_train, pickups_train,'yo', alpha=.4, label='Training Data')
plt.title('Taxi Pickups vs Time of Day (Full Dataset)')
plt.xlabel('Time of Day in Minutes')
plt.ylabel('Number of Taxi Pickups')

# Plot OLS Line
x1 = [0, max(time)]
y1 = [beta0, beta0 + max(time)*beta1]
plt.plot(x1, y1, 'k-', lw=3, label='OLS Line')

# Show Plot
plt.legend()
plt.show()
````

**3.3**

````python
## Code here

# Training R Squared
train_r2 = results_sm.rsquared

print('R Squared for the training model:', round(train_r2, 3))
````

````python
test_mean = np.mean(taxi_test.PickupCount)

avg_errors, model_errors = [0]*250, [0]*250

for i in range(len(taxi_test)):
    avg_errors[i] = (taxi_test.PickupCount.values[i] - test_mean)**2
    model_errors[i] = (beta0 + beta1*taxi_test.TimeMin.values[i] - taxi_test.PickupCount.values[i])**2

test_r2 = 1 - sum(model_errors)/sum(avg_errors)

print('R Squared for the test set:', round(test_r2, 3))
````
